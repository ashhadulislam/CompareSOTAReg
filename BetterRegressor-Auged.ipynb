```python
import lib
import numpy as np
import os
from sklearn.model_selection import train_test_split
```

```python
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor


```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, \
    mean_squared_log_error, r2_score, explained_variance_score, \
    median_absolute_error
```

```python
import pandas as pd
```

```python
file_loc="../selected-tabular-datasets/"

```

```python
num_nbrs_list=[3,4,5]
```

```python
os.listdir(file_loc)
num_nbrs=3
```

```python
files=os.listdir(file_loc)
files.sort()
```

```python
files
```

```python
res={}
res['dataset']=[]
res['dataset_shape']=[]
res['dataset_aug_shape']=[]
res['model']=[]
res['num_nbrs']=[]
res['mae_auged']=[]
res['mse_auged']=[]
res['rmse_auged']=[]
res['rsq_auged']=[]
res['expvar_auged']=[]
res['median_abs_err_auged']=[]



for f in files:
    if "housing" in f:
        continue

    if "ANACALT" in f:
        continue

    if "concrete" in f:
        continue
    print(f)
    df=pd.read_csv(file_loc+"/"+f)
    # print(df.head())
    X=df.iloc[:,:-1].values
    y=df.iloc[:,-1].values
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.33, random_state=42)

    print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
    for num_nbrs in num_nbrs_list:
        X_train_aug,y_train_aug=lib.oversample_data(X_train,y_train,num_nbrs=num_nbrs)
        print("Bef:",X_train.shape,y_train.shape)
        print("Aft:",X_train_aug.shape,y_train_aug.shape)
    
    
        models = {
            "cbr":CatBoostRegressor(),
            "xgbr":XGBRegressor(),
            "lgbmr":LGBMRegressor()
        }
        for k,model in models.items():
            print("Model",k)
            
            model.fit(X_train_aug,y_train_aug)
            print("Training complete")
            predicted=model.predict(X_test)
            actual=y_test
            
            
        
            # Mean Absolute Error (MAE)
            mae = mean_absolute_error(actual, predicted)
            print("Mean Absolute Error (MAE):", mae)
            
            # Mean Squared Error (MSE)
            mse = mean_squared_error(actual, predicted)
            print("Mean Squared Error (MSE):", mse)
            
            # Root Mean Squared Error (RMSE)
            rmse = np.sqrt(mse)
            print("Root Mean Squared Error (RMSE):", rmse)
            
            # R-squared (R2)
            r2 = r2_score(actual, predicted)
            print("R-squared (R2):", r2)
            
            # Explained Variance Score
            explained_variance = explained_variance_score(actual, predicted)
            print("Explained Variance Score:", explained_variance)
            
            # Median Absolute Error
            median_ae = median_absolute_error(actual, predicted)
            print("Median Absolute Error:", median_ae)
        
            res['dataset'].append(f)
            res['dataset_shape'].append(str(X_train.shape[0]))
            res['dataset_aug_shape'].append(str(X_train_aug.shape[0]))      
            res['model'].append(k)
            res['num_nbrs'].append(num_nbrs)
            res['mae_auged'].append(mae)
            res['mse_auged'].append(mse)
            res['rmse_auged'].append(rmse)
            res['rsq_auged'].append(r2)
            res['expvar_auged'].append(explained_variance)
            res['median_abs_err_auged'].append(median_ae)
```

```python
res_df=pd.DataFrame(res)
```
