```python
import lib
import numpy as np
import os
from sklearn.model_selection import train_test_split
```

```python
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor


```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, \
    mean_squared_log_error, r2_score, explained_variance_score, \
    median_absolute_error
```

```python
import pandas as pd
```

```python
file_loc="../selected-tabular-datasets/"

```

```python
os.listdir(file_loc)
```

```python
res={}
res['dataset']=[]
res['model']=[]
res['mae_original']=[]
res['mse_original']=[]
res['rmse_original']=[]
res['rsq_original']=[]
res['expvar_original']=[]
res['median_abs_err_original']=[]



for f in os.listdir(file_loc):
    if "housing" in f:
        continue

    if "ANACALT" in f:
        continue

    if "concrete" in f:
        continue

    print(f)
    df=pd.read_csv(file_loc+"/"+f)
    # print(df.head())
    X=df.iloc[:,:-1].values
    y=df.iloc[:,-1].values
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.33, random_state=42)

    print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)


    models = {
        "cbr":CatBoostRegressor(),
        "xgbr":XGBRegressor(),
        "lgbmr":LGBMRegressor()
    }
    for k,model in models.items():
        print("Model",k)
        res['model'].append(k)
        model.fit(X_train,y_train)
        print("Training complete")
        predicted=model.predict(X_test)
        actual=y_test
        
        res['dataset'].append(f)
    
        # Mean Absolute Error (MAE)
        mae = mean_absolute_error(actual, predicted)
        print("Mean Absolute Error (MAE):", mae)
        
        # Mean Squared Error (MSE)
        mse = mean_squared_error(actual, predicted)
        print("Mean Squared Error (MSE):", mse)
        
        # Root Mean Squared Error (RMSE)
        rmse = np.sqrt(mse)
        print("Root Mean Squared Error (RMSE):", rmse)
        
        # R-squared (R2)
        r2 = r2_score(actual, predicted)
        print("R-squared (R2):", r2)
        
        # Explained Variance Score
        explained_variance = explained_variance_score(actual, predicted)
        print("Explained Variance Score:", explained_variance)
        
        # Median Absolute Error
        median_ae = median_absolute_error(actual, predicted)
        print("Median Absolute Error:", median_ae)
    
    
        res['mae_original'].append(mae)
        res['mse_original'].append(mse)
        res['rmse_original'].append(rmse)
        res['rsq_original'].append(r2)
        res['expvar_original'].append(explained_variance)
        res['median_abs_err_original'].append(median_ae)
```

```python
# res
```

```python
res_df=pd.DataFrame(res)
```
