```python
import lib
import numpy as np
import os
from sklearn.model_selection import train_test_split

```

```python
import pycaret
pycaret.__version__
# import pycaret regression and init setup
from pycaret.regression import RegressionExperiment
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, \
    mean_squared_log_error, r2_score, explained_variance_score, \
    median_absolute_error
import pandas as pd
```

```python
file_loc="../datasets/selected_datasets/"
num_nbrs_list=[3,4,5]
os.listdir(file_loc)
num_nbrs=3
```

```python
files=os.listdir(file_loc)
files.sort()
```

```python




for f in files:
    if "housing" in f:
        continue

    if "ANACALT" in f:
        continue

    if "concrete" in f:
        continue
    print(f)
    df=pd.read_csv(file_loc+"/"+f)
    # print(df.head())
    X=df.iloc[:,:-1].values
    y=df.iloc[:,-1].values
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.33, random_state=42)

    print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)


    X_train_aug,y_train_aug=X_train,y_train
    print("Bef:",X_train.shape,y_train.shape)
    print("Aft:",X_train_aug.shape,y_train_aug.shape)
    s = RegressionExperiment()
    s.setup(np.concatenate((X_train_aug,y_train_aug.reshape(-1,1)),axis=1),
              test_data=np.concatenate((X_test,y_test.reshape(-1,1)),axis=1),
            index=False,
            verbose=False,
            session_id = 123)
#         print("COmparing models")
    best = s.compare_models(verbose=False)
#         print("Getting leaderboard")
    compare_df=s.get_leaderboard(verbose=False)
    compare_df_sorted=compare_df.sort_values(by=list(compare_df.columns)[2:])
    compare_df_sorted=compare_df_sorted.drop_duplicates(subset=["Model Name"])
    compare_df_sorted.to_csv(f"pyCaretResults/no_aug/{f}",index=False)
    
```

```python
models_df=s.models()
```
