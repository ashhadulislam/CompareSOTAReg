```python
import pandas as pd
```

```python
df_original=pd.read_csv("SOTARegOnOriginal.csv")
```

```python
df_knnor=pd.read_csv("SOTARegOnKNNOR.csv")
```

```python
len(df_original.dataset.unique())
```

```python
modelwise_stats={}
win_count=0
loss_count=0
for f in df_original.dataset.unique():
    print(f)
    df_original_f=df_original[df_original["dataset"]==f]
    # print(df_original_f.shape)

    df_knnor_f=df_knnor[df_knnor["dataset"]==f]
    # print(df_knnor_f.shape)

    for model in df_knnor_f.model.unique():
        print(model)
        if model not in modelwise_stats:
            modelwise_stats[model]={}
            
        df_knnor_f_m=df_knnor_f[df_knnor_f["model"]==model]
        df_original_f_m=df_original_f[df_original_f["model"]==model]
        

        print(df_knnor_f_m.shape)
        print(df_original_f_m.shape)
        for metric in ["mae","mse","rmse","rsq","expvar","median_abs_err"]:
            if metric not in modelwise_stats[model]:
                modelwise_stats[model][metric]={}
                modelwise_stats[model][metric]["win_count"]=0
                modelwise_stats[model][metric]["loss_count"]=0

            if min(list(df_knnor_f_m[metric+"_auged"]))<list(df_original_f_m[metric+"_original"])[0]:
                print("Win",min(list(df_knnor_f_m[metric+"_auged"])),list(df_original_f_m[metric+"_original"])[0])
                win_count+=1
                modelwise_stats[model][metric]["win_count"]+=1
            else:
                print("Loss",min(list(df_knnor_f_m[metric+"_auged"])),list(df_original_f_m[metric+"_original"])[0])
                loss_count+=1
                modelwise_stats[model][metric]["loss_count"]+=1
                

        


        # for metric in ["mae","mse","rmse","rsq","expvar","median_abs_err"]:
        # for metric in ["rsq","expvar","median_abs_err"]:            
        #     if list(df_knnor_f_m[metric+"_auged"])[0]>=list(df_original_f_m[metric+"_original"])[0]:
        #         print("Win")
        #         win_count+=1
        #     else:
        #         loss_count+=1
        #         print("Loss")
    
```

```python
def median(a):
    
    if a == []:
        
        return None
    
    sorted_l = sorted(a)
        
    
    if len(sorted_l) % 2 != 0:
        
        indexx = ((len(sorted_l) / 2) - 0.5) 
        
        indexx = int(indexx)
        
        answer = sorted_l[indexx]        
    
    else:                
            
        length_l = len(sorted_l)
        
        
        first_index = int(length_l /2) - 1
        
        second_index = int(first_index + 1)
        
        answer = (sorted_l[first_index] + sorted_l[second_index]) / 2
            
    return answer
```
