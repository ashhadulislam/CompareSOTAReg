```python
import pandas as pd
import os

```

```python
df_reference=pd.read_csv("pyCaretResults/after_aug/nbr_3_bank8FM.csv")
```

```python
df_reference.head()
```

```python
model_names=list(df_reference["Model Name"].unique())
metrics=list(df_reference.columns)[2:]
if "Dummy Regressor" in model_names:
    model_names.remove("Dummy Regressor")
model_names.sort()    
```

```python
print(model_names,metrics)
```

```python
aug_list_files= os.listdir("pyCaretResults/after_aug")
aug_list_files.sort()
```

```python
non_aug_list_files=os.listdir("pyCaretResults/no_aug")
non_aug_list_files.sort()
```

```python
result_data_dic={}

result_data_dic["model"]=[]
result_data_dic["metrics"]=[]
result_data_dic["wins"]=[]
result_data_dic["losses"]=[]
```

```python
for model_name in model_names:
    
    for metric in metrics:
        result_data_dic["model"].append(model_name)
        result_data_dic["metrics"].append(metric)
        result_data_dic["wins"].append(0)
        result_data_dic["losses"].append(0)
        
        for non_aug_file in non_aug_list_files:

            print(non_aug_file,model_name,metric)
            df_non_aug_file=pd.read_csv("pyCaretResults/no_aug/"+non_aug_file)
            df_non_aug_file_model=df_non_aug_file[df_non_aug_file["Model Name"]==model_name]
#             print(df_non_aug_file_model.shape)
            unaug_metric_value=list(df_non_aug_file_model[metric])[0]
            print("non_aug",metric,unaug_metric_value)
            
            
            num_nbrs=[3,4,5]
            metric_vals=[]
            for nbr_num in num_nbrs:
                aug_file_name=f"nbr_{nbr_num}_{non_aug_file}"
        #         print(aug_file_name)

                if aug_file_name in aug_list_files:
#                     print(aug_file_name)
                    df_augmented=pd.read_csv("pyCaretResults/after_aug/"+aug_file_name)
                    df_augmented_model=df_augmented[df_augmented["Model Name"]==model_name]
                    aug_metric_value=list(df_augmented_model[metric])[0]
#                     print(f'aug {nbr_num} {aug_metric_value}')
                    metric_vals.append(aug_metric_value)
            
            print("Aug",metric_vals)
            if metric!='R2':
                best_aug_metric=min(metric_vals)
                if best_aug_metric<=unaug_metric_value:
                    result_data_dic["wins"][-1]+=1
                else:
                    result_data_dic["losses"][-1]+=1
            else:                
                best_aug_metric=max(metric_vals)
                if best_aug_metric>=unaug_metric_value:
                    result_data_dic["wins"][-1]+=1
                else:
                    result_data_dic["losses"][-1]+=1
                
                
            print(metric,"unaug",unaug_metric_value,"aug",best_aug_metric)
            print("*"*10)
            
#         break
#     break
    
```

```python
result_data_dic_df=pd.DataFrame(result_data_dic)
```

```python
result_data_dic_df
```

```python
result_data_dic_df.to_csv("pyCaretResults/comparison.csv",index=False)
```

```python
result_data_dic_df.model.unique()
```

```python
model_agg={}
model_agg["model"]=[]
model_agg["total_wins"]=[]
model_agg["total_losses"]=[]
model_agg["percentage"]=[]
for model_name in result_data_dic_df.model.unique():
    result_data_dic_df_model=result_data_dic_df[result_data_dic_df["model"]==model_name]
    total_wins=result_data_dic_df_model["wins"].sum()
    total_losses=result_data_dic_df_model["losses"].sum()    
    percentage=round(total_wins/(total_wins+total_losses)*100,2)
    
    model_agg["model"].append(model_name)
    model_agg["total_wins"].append(total_wins)
    model_agg["total_losses"].append(total_losses)
    model_agg["percentage"].append(percentage)
    print(model_name,result_data_dic_df_model.shape)
```

```python
model_agg_df=pd.DataFrame(model_agg)
```

```python
model_agg_df=model_agg_df.sort_values(by=["percentage"],ascending=False)
```

```python
model_agg_df.to_csv("pyCaretResults/comparison_aggregated_onMetrics.csv",index=False)


```
